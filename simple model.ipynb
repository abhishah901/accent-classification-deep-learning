```python
import tensorflow as tf
# from tf.keras.optimizers import SGD
import sklearn
 
# np.random.seed(1337)  # for reproducibility
# import tf.keras.preprocessing.sequence # import sequence
# import tf.keras.utils.np_utils # import np_utils
# import tf.keras.models.Sequential # import Sequential
# import tf.keras.layers.core.Dense # import Dense, Dropout, Activation
# import tf.keras.layers.core.Dropout
# import tf.keras.layers.core.Activation
# import tf.keras.layers.recurrent.LSTM # import LSTM
```

```python
test_dim = 2999
maxlen = 100
batch_size = 30
nb_filter = 64
filter_length_1 = 50
filter_length_2 = 25
hidden_dims = 250
nb_epoch = 8
nb_classes = 3
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
model = tf.keras.models.Sequential()
# model = tf.expand_dims(model, axis=-1)
model.add(tf.keras.layers.LSTM(32, return_sequences=True, stateful=False))
model.add(tf.keras.layers.LSTM(20))

model.add(tf.keras.layers.Dense(nb_classes, activation='softmax'))


model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=["accuracy"])

print("Train...")
model.fit(X_train, y_train, batch_size=batch_size, nb_epoch=5, validation_data=(X_test, y_test))
```

```python
y_pred=model.predict_classes(X_test, batch_size=batch_size)
print(classification_report(y_test, y_pred))
```

```python
classifier = tf.keras.models.Sequential()
classifier.add(tf.keras.layers.Dense(units=6, kernel_initializer='glorot_uniform',activation='relu'))

#adding second and third hidden layer
classifier.add(tf.keras.layers.Dense(units=6, kernel_initializer='glorot_uniform',activation='relu'))
classifier.add(tf.keras.layers.Dense(units=6, kernel_initializer='glorot_uniform',activation='relu'))


#adding output layer
classifier.add(tf.keras.layers.Dense(units=1, kernel_initializer='glorot_uniform',activation='sigmoid'))

#compiling the ANN
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

#fitting the ANN to traing set
classifier.fit(X_train, y_train, batch_size=20, epochs=500)

#prediction
y_pred = classifier.predict(X_test)
y_pred =(y_pred > 0.5)
```
